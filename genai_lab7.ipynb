Summarize long texts using a pre-trained summarization model using Hugging face model. Load the summarization pipeline. Take a passage as input and obtain the summarized text.

```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
def summarize_text(text, max_length=80, min_length=40):
    summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=False)
    return summary[0]["summary_text"]
```

```python
text = """
The Tamil language is among the oldest and most revered languages in the world. It is a Dravidian language, deeply rooted in history and tradition, and serves as the official language of Tamil Nadu, a vibrant state in South India, and Puducherry. Beyond India, Tamil is an essential part of the cultural and linguistic identity of communities in Sri Lanka, Singapore, and Malaysia, as well as among Tamil-speaking diaspora around the globe.

Tamil boasts a literary heritage that spans over two millennia, with its origins tracing back to the Sangam period. The Sangam-era literature, consisting of poems and epics, reflects the intellectual depth and cultural richness of ancient Tamil society. Renowned works like the Thirukkural, written by the great poet and philosopher Thiruvalluvar, continue to inspire readers with their timeless wisdom. Masterpieces such as Silappatikaram and Manimekalai, celebrated epics, highlight Tamil's poetic brilliance and artistic storytelling.

The uniqueness of Tamil lies in its ability to bridge the past and the present. Despite its ancient origins, Tamil is a living language that thrives in modern times. It has seamlessly adapted to contemporary fields like education, technology, science, and entertainment while retaining its cultural and linguistic essence. Tamil cinema, for example, plays a significant role in popular culture and continues to promote the language on global platforms.

Tamil is more than just a language; it is a symbol of identity, pride, and heritage for millions of speakers worldwide. It is celebrated not only for its linguistic elegance but also for the way it encapsulates the cultural ethos, traditions, and values of its people.
"""
```

```python
summary = summarize_text(text)
```

```python
print("Original Text:")
print(text)
print("\nSummarized Text:")
print(summary)
```
